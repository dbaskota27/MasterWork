```python
# Enhanced Trading Journal Dashboard in Jupyter Notebook (Updated & Fixed)
# Fixes: CSV parsing errors, NaN in SelectMultiple widget, date NaT issues

# Install if missing: !pip install pandas plotly ipywidgets tabulate

import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from ipywidgets import interact, Tab, Output, Textarea, Button
from IPython.display import display, clear_output
import json
import os
import numpy as np
from tabulate import tabulate

# File paths
CSV_FILE = 'jan.csv'     # ← Change this to your exported CSV path
NOTES_FILE = 'notes.json'

# ────────────────────────────────────────────────────────────────
# Load and clean data with robust parsing
# ────────────────────────────────────────────────────────────────
print("Loading CSV...")

try:
    df = pd.read_csv(
        CSV_FILE,
        on_bad_lines='warn',           # Warn and skip bad rows (Python 3.9+)
        encoding='utf-8',
        quoting=1,                     # csv.QUOTE_ALL - more forgiving
        doublequote=True
    )
    print("CSV loaded successfully (some bad rows may have been skipped).")
except Exception as e:
    print(f"CSV loading failed: {e}")
    print("Try opening the file in a text editor and check line 757 for extra commas.")
    raise

# Parse dates
df['Activity Date'] = pd.to_datetime(df['Activity Date'], errors='coerce')

# Clean Amount column
df['Amount'] = pd.to_numeric(
    df['Amount'].astype(str).str.replace('[$(),]', '', regex=True),
    errors='coerce'
)

# Sort by date
df = df.sort_values('Activity Date').reset_index(drop=True)

print(f"Rows loaded: {len(df)}")
print(f"Rows with missing Activity Date: {df['Activity Date'].isna().sum()}")
print(f"Rows with missing Amount: {df['Amount'].isna().sum()}")

# ────────────────────────────────────────────────────────────────
# Extended metrics function
# ────────────────────────────────────────────────────────────────
def calculate_metrics(filtered_df):
    total_pl = filtered_df['Amount'].sum()
    total_trades = len(filtered_df)
    wins = filtered_df[filtered_df['Amount'] > 0]
    losses = filtered_df[filtered_df['Amount'] < 0]
    win_count = len(wins)
    loss_count = len(losses)
    win_rate = (win_count / total_trades * 100) if total_trades > 0 else 0
    avg_win = wins['Amount'].mean() if win_count > 0 else 0
    avg_loss = losses['Amount'].mean() if loss_count > 0 else 0
    profit_factor = abs(wins['Amount'].sum() / losses['Amount'].sum()) if loss_count > 0 else 0
    filtered_df['Cum PL'] = filtered_df['Amount'].cumsum()
    max_drawdown = (filtered_df['Cum PL'] - filtered_df['Cum PL'].cummax()).min()
    expectancy = (win_rate / 100 * avg_win) + ((1 - win_rate / 100) * avg_loss)
    total_volume = filtered_df['Quantity'].sum() if 'Quantity' in filtered_df else 0

    return {
        'Total P/L': f"${total_pl:,.2f}",
        'Total Trades': total_trades,
        'Win Rate (%)': f"{win_rate:.2f}",
        'Avg Win': f"${avg_win:,.2f}",
        'Avg Loss': f"${avg_loss:,.2f}",
        'Profit Factor': f"{profit_factor:.2f}",
        'Max Drawdown': f"${max_drawdown:,.2f}",
        'Expectancy': f"${expectancy:,.2f}",
        'Total Volume': total_volume
    }

# ────────────────────────────────────────────────────────────────
# Group trades to estimate closed P/L
# ────────────────────────────────────────────────────────────────
def group_trades(filtered_df):
    if 'Description' not in filtered_df or 'Trans Code' not in filtered_df:
        return pd.DataFrame()
    grouped = filtered_df.groupby('Description')
    closed = []
    for name, group in grouped:
        bto_sum = group[group['Trans Code'] == 'BTO']['Amount'].sum()
        stc_sum = group[group['Trans Code'] == 'STC']['Amount'].sum()
        if bto_sum != 0 and stc_sum != 0:
            net_pl = stc_sum + bto_sum
            closed.append({'Description': name, 'Net P/L': net_pl})
    return pd.DataFrame(closed)

# ────────────────────────────────────────────────────────────────
# Notes
# ────────────────────────────────────────────────────────────────
def load_notes():
    if os.path.exists(NOTES_FILE):
        with open(NOTES_FILE, 'r') as f:
            return json.load(f)
    return {}

def save_note(trade_index, note):
    notes = load_notes()
    notes[str(trade_index)] = note
    with open(NOTES_FILE, 'w') as f:
        json.dump(notes, f, indent=2)

# ────────────────────────────────────────────────────────────────
# Clean instruments (remove NaN/None/empty)
# ────────────────────────────────────────────────────────────────
if 'Instrument' in df:
    instruments_raw = df['Instrument'].dropna().unique()
    instruments = [str(x).strip() for x in instruments_raw if pd.notna(x) and str(x).strip()]
    if not instruments:
        instruments = ['(No valid instruments)']
        print("Warning: No valid Instrument values found.")
else:
    instruments = ['(Instrument column missing)']
    print("Warning: 'Instrument' column not found in data.")

# ────────────────────────────────────────────────────────────────
# Widgets
# ────────────────────────────────────────────────────────────────
instrument_widget = widgets.SelectMultiple(
    options=instruments,
    value=instruments if instruments != ['(No valid instruments)'] else [],
    description='Instruments:',
    rows=8,
    disabled=False
)

min_date = df['Activity Date'].min()
max_date = df['Activity Date'].max()

start_date_widget = widgets.DatePicker(
    value=min_date.date() if pd.notna(min_date) else None,
    description='Start Date:'
)

end_date_widget = widgets.DatePicker(
    value=max_date.date() if pd.notna(max_date) else None,
    description='End Date:'
)

filter_box = widgets.VBox([instrument_widget, start_date_widget, end_date_widget])

# ────────────────────────────────────────────────────────────────
# Tabs
# ────────────────────────────────────────────────────────────────
tab = Tab()
tab.children = [Output() for _ in range(4)]
tab.titles = ('Metrics', 'Trades Table', 'Charts', 'Notes')

# ────────────────────────────────────────────────────────────────
# Update function
# ────────────────────────────────────────────────────────────────
def update_dashboard(change=None):
    selected = list(instrument_widget.value)
    start = start_date_widget.value
    end = end_date_widget.value

    mask = (
        df['Instrument'].astype(str).isin(selected) &
        (df['Activity Date'].dt.date >= (start or min_date.date())) &
        (df['Activity Date'].dt.date <= (end or max_date.date()))
    )
    filtered = df[mask].copy()

    # Tab 0: Metrics
    with tab.children[0]:
        clear_output(wait=True)
        if filtered.empty:
            display(widgets.HTML("<h3>No data matches filters</h3>"))
        else:
            metrics = calculate_metrics(filtered)
            metrics_html = tabulate([metrics], headers='keys', tablefmt='html')
            display(widgets.HTML(metrics_html))

    # Tab 1: Trades Table
    with tab.children[1]:
        clear_output(wait=True)
        display(filtered)

    # Tab 2: Charts
    with tab.children[2]:
        clear_output(wait=True)
        if filtered.empty:
            display(widgets.HTML("<h3>No data for charts</h3>"))
            return

        filtered['Cum PL'] = filtered['Amount'].cumsum()
        fig1 = px.line(filtered, x='Activity Date', y='Cum PL', title='Equity Curve')
        fig1.show()

        pl_inst = filtered.groupby('Instrument', as_index=False)['Amount'].sum()
        fig2 = px.bar(pl_inst, x='Instrument', y='Amount', title='P/L by Instrument')
        fig2.show()

        filtered['Type'] = filtered['Description'].str.contains('Put', case=False, na=False).map({True: 'Put', False: 'Call'})
        type_cnt = filtered['Type'].value_counts().reset_index()
        fig3 = px.pie(type_cnt, values='count', names='Type', title='Put vs Call')
        fig3.show()

        closed = group_trades(filtered)
        if not closed.empty:
            fig4 = px.histogram(closed, x='Net P/L', title='Closed Trades P/L Distribution')
            fig4.show()

    # Tab 3: Notes
    with tab.children[3]:
        clear_output(wait=True)
        notes = load_notes()
        idx_input = widgets.IntText(value=0, description='Trade Index:')
        note_area = Textarea(value=notes.get(str(idx_input.value), ''), description='Note:', rows=6)
        save_btn = Button(description='Save Note')

        def on_save(b):
            save_note(idx_input.value, note_area.value)
            print(f"Note saved for index {idx_input.value}")

        save_btn.on_click(on_save)
        display(widgets.VBox([idx_input, note_area, save_btn]))
        print("Current notes:", notes)

# Bind updates
instrument_widget.observe(update_dashboard, names='value')
start_date_widget.observe(update_dashboard, names='value')
end_date_widget.observe(update_dashboard, names='value')

# Show dashboard
display(filter_box)
display(tab)
update_dashboard()
```
